* As a starting point, we analyze two fundamental feature-extraction methods: Root Mean Square Energy (RMSE) and Zero Crossing Rate (ZCR). These metrics are essential in acoustic signal analysis, especially for sound classification, event detection, and speech recognition.

* **RMSE** measures the average energy of an acoustic signal by taking the square root of the mean of the squared amplitudes. This measure is especially useful for evaluating sound intensity and distinguishing between high- and low-amplitude signals.

* **ZCR** measures how often the amplitude of the signal changes sign (crosses the time axis). It is a crucial descriptor for identifying transient or fricative sounds, since signals with a high ZCR tend to contain more high-frequency noise or transient components.

* Both metrics are standard in the audio-processing literature and, when combined with other acoustic descriptors, significantly improve the accuracy and robustness of predictive classification models.

### Zero Crossing Rate (ZCR) with Librosa

```python
zcr_audio01 = librosa.feature.zero_crossing_rate(y=audio01, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
zcr_audio02 = librosa.feature.zero_crossing_rate(y=audio02, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

As seen in the blue graph for audio02 (sad), the ZCR appears to be more present. This indicates a greater amount of transients or noise, but it does not let us intuitively characterize or differentiate the tone of voice between the two samples. For that reason, it is clear that this feature is more useful when providing complementary information together with techniques such as RMSE, MFCCs, etc.
